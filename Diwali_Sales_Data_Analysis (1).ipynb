# Diwali Sales Data Analysis
This notebook performs Exploratory Data Analysis (EDA) and extracts insights from the Diwali Sales dataset.

```python
# Load and display the first few rows of the dataset
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'arbitrary_sales_data.csv'  # Replace with the actual file path
# Load the CSV file into a Pandas DataFrame for data manipulationsales_data = pd.read_csv(file_path)

# Display the first few rows of the dataset
# Display the first few rows of the dataset to preview the datasales_data.head()
```

## Data Overview
Basic statistics and structure of the data.

```python
# Check for missing values
# Generate summary statistics for numerical columns
# Pandas .info() provides a concise summary of the DataFrame structuresales_data.info()

# Summary statistics
# Pandas .describe() generates descriptive statistics for numerical columnssales_data.describe()
```

## Exploratory Data Analysis (EDA)
### Sales Trends and Seasonality

```python
# Plot sales amount over time
# Convert the Purchase_Date column to a datetime format for easier manipulationsales_data['Purchase_Date'] = pd.to_datetime(sales_data['Purchase_Date'])
sales_trend = sales_data.groupby('Purchase_Date')['Sales_Amount'].sum().reset_index()

# Matplotlib is used to configure the figure size for the plotplt.figure(figsize=(12, 6))
# Seaborn lineplot is used to visualize sales trends over timesns.lineplot(data=sales_trend, x='Purchase_Date', y='Sales_Amount')
plt.title('Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Sales Amount')
# Display the plot generated by Matplotlib or Seabornplt.show()
```

### Customer Segmentation
Segregate data by gender, age group, location, and product category.

```python
# Sales distribution by gender
# Matplotlib is used to configure the figure size for the plotplt.figure(figsize=(8, 6))
# Create a bar plot to show total sales by gender using Seabornsns.barplot(data=sales_data, x='Gender', y='Sales_Amount', estimator=sum, ci=None)
plt.title('Total Sales by Gender')
plt.xlabel('Gender')
plt.ylabel('Total Sales')
# Display the plot generated by Matplotlib or Seabornplt.show()

# Sales by age group
# Matplotlib is used to configure the figure size for the plotplt.figure(figsize=(10, 6))
# Create a bar plot to show total sales by age group using Seabornsns.barplot(data=sales_data, x='Age_Group', y='Sales_Amount', estimator=sum, ci=None)
plt.title('Total Sales by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Sales')
# Display the plot generated by Matplotlib or Seabornplt.show()
```

### Evaluating Significant Factors
Impact of pricing strategies and advertisements.

```python
# Sales distribution by pricing strategy
# Matplotlib is used to configure the figure size for the plotplt.figure(figsize=(10, 6))
# Create a bar plot to evaluate total sales by pricing strategy using Seabornsns.barplot(data=sales_data, x='Pricing_Strategy', y='Sales_Amount', estimator=sum, ci=None)
plt.title('Total Sales by Pricing Strategy')
plt.xlabel('Pricing Strategy')
plt.ylabel('Total Sales')
# Display the plot generated by Matplotlib or Seabornplt.show()

# Sales distribution by advertisement channel
# Matplotlib is used to configure the figure size for the plotplt.figure(figsize=(12, 6))
# Create a bar plot to analyze total sales by advertisement channel using Seabornsns.barplot(data=sales_data, x='Advertisement_Channel', y='Sales_Amount', estimator=sum, ci=None)
plt.title('Total Sales by Advertisement Channel')
plt.xlabel('Advertisement Channel')
plt.ylabel('Total Sales')
# Display the plot generated by Matplotlib or Seabornplt.show()
```
